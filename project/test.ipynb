```python
import numpy as np

import game2048
import pygame
import gym
```

```python
env = gym.make('envs/game2048-v0',reward_type=2, render_mode='human')
observation = env.reset()
```

```python
env.env.board = np.array([[2, 2 ,4 ,2], [5, 6, 7, 8], [9, 10, 11, 12],[5, 6, 7, 8]])

for _ in range(10000):
    env.render()
    events = pygame.event.get()
    terminated = False
    info = {}
    for event in events:
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_DOWN:
                observation, reward, terminated, info = env.step(3)
            if event.key == pygame.K_UP:
                observation, reward, terminated, info = env.step(1)
            if event.key == pygame.K_LEFT:
                observation, reward, terminated, info = env.step(2)
            if event.key == pygame.K_RIGHT:
                observation, reward, terminated, info = env.step(0)
    print(f"Reward: {info.get('reward')}")
    print(f"observation: {observation}")
    if terminated:
        print(f"reward: {reward}")
        print(info.get('board'))
    if terminated:
        observation = env.reset()
    env.render()
```
